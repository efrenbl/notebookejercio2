```python
#Importando pyspark
from pyspark import SparkContext
from pyspark.sql import *
from pyspark.sql.functions import asc, desc, monotonically_increasing_id, col, ltrim, rtrim
```

```python
#Creaci√≥n de  dataframe
df = (spark
  .read                                              
  .option("inferSchema","true")                 
  .option("header","true")                           
  .csv("demo.csv"))
```

```python
#Imprimiendo Esquema de datos
df.printSchema()
```

```python
#mostrando dataframe
df.show()
```

```python
#limpiando name
df=df.where("name!='Unknown'")
```

```python
#limpiando country
df=df.where("country!='null'")
```

```python
 df = df.withColumn('country', ltrim(df.country)) # recortar el espacio en blanco izquierdo de la columna country
```

```python
 df = df.withColumn('country', rtrim(df.country)) # recortar el espacio en blanco derecho de la columna country
```

```python
#mostrando resultado de limpieza
df.show()
```

```python
#contando registros/filas
df.count()
```

```python
#guardando cantidas para posterior uso
a =  df.count()
```

```python
print(a)
```

```python
#creando dt nuevo
dfdict = df.select("country")
```

```python
#eliminado duplicados
dfdict = dfdict.drop_duplicates(subset = ['country'])
```

```python
#contando registros/filas
dfdict.count()
```

```python
#mostrando resultado
dfdict.show()
```

```python
#ordenando columna country
dfdict= dfdict.orderBy(asc("country"))
```
